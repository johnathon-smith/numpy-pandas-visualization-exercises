### Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(user, host, password, database):
    url = f"mysql+pymysql://{user}:{password}@{host}/{database}"
    return url
```

```python
from env import host, user, password
chipotle_url = get_db_url(user, host, password, 'chipotle')
chipotle_query = "SELECT * FROM orders"
chipotle = pd.read_sql(chipotle_query, chipotle_url)
chipotle
```

```python
#Convert the price string to a numeric value
chipotle.item_price = chipotle.item_price.str.replace('$', '').astype(float)
chipotle
```

```python
chipotle_popular_items = chipotle.groupby('item_name').quantity.sum().sort_values().tail(4)
```

```python
popular_item_revenue = chipotle[chipotle.item_name.isin(chipotle_popular_items.index)].groupby('item_name').item_price.sum()
popular_item_revenue = popular_item_revenue.sort_values(ascending = False).reset_index()
popular_item_revenue
```

```python
sns.barplot(data = popular_item_revenue, x = 'item_price', y = 'item_name')
plt.xlabel('Revenue ($)')
plt.ylabel('Item Name')
plt.title('Revenue From Most Popular Items')
plt.show()
```
