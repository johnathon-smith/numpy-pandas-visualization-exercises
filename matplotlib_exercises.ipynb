### Use matplotlib to plot the following equation:

### y = x**2 − x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = list(range(1,101))
```

```python
x = pd.Series(x)
x
```

```python
y = x**2 - x + 2
y
```

```python
plt.figure(figsize=(14,8))
plt.plot(x,y)
plt.annotate('Origin', xy = (0,0), xytext = (0,2000), arrowprops={'facecolor':'blue'}, fontsize = 20)
plt.show()
```

```python
#Create a series for x with values from 1 to 10
x = list(range(1, 6))
x = pd.Series(x)
x
```

```python
#Create my y-values
y1 = pow(x, 0.5)
y2 = pow(x,3)
y3 = pow(2,x)
y4 = 1 / (x + 1)
```

```python
#Create subplots
n_rows = 2
n_cols = 2

plt.subplot(221)
plt.plot(x,y1)
plt.title('$y = \sqrt{x}$')

plt.subplot(222)
plt.plot(x,y2)
plt.title('$y = x^3$')

plt.subplot(223)
plt.plot(x,y3)
plt.title('$y = 2^x$')

plt.subplot(224)
plt.plot(x,y4)
plt.title('y = 1 / (x + 1)')

plt.tight_layout()
plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(10, 10))
plt.scatter(x,y1, c = 'red')
plt.plot(x,y1, c = 'red', label = '$y = \sqrt{x}$')
plt.scatter(x,y2, c = 'green')
plt.plot(x,y2, c = 'green', label = '$y = x^3$')
plt.scatter(x,y3, c = 'purple')
plt.plot(x,y3, c = 'purple', label = '$y = 2^x$')
plt.scatter(x,y4, c = 'blue')
plt.plot(x,y4, c = 'blue', label = '$y = 1 / (x + 1)$')

plt.legend(loc='best')
plt.title('Four Functions')
plt.xlabel('$x$')

plt.show()
```
