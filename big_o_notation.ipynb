```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#The following is needed to use a factorial funtion on a pandas series
import scipy.special
```

### Curves to graph
    - y1 = 0n + 1 and label the curve "O1"
    - y2 = log(n) and label the curve "O(log n)"
    - y3 = n and label the curve "O(n)"
    - y4 = n * log(n) and label it "O(n log n)"
    - y5 = n**2 and label it "O(n^2)"
    - y6 = 2**n and label it "O(2^n)"
    - y7 = n! and label it "O(n!)"
    - y8 = n**n and label it "O(n^n)"
### Make sure to:
    - Title your chart "Big O Notation"
    - Label your x axis "Elements"
    - Label your y axis "Operations"
    - Label your curves or make a legend for the curves
    - Use LaTex notation where possible

```python
#Create n values
n = list(range(1,4))
n = pd.Series(n)
```

```python
#Create y values
y1 = 0*n + 1
y2 = np.log(n)
y3 = n
y4 = n * np.log(n)
y5 = n**2
y6 = 2**n
y7 = scipy.special.factorial(n)
y8 = n**n
```

```python
#Begin plotting the curves
plt.figure(figsize = (14,12))

plt.scatter(n,y1, c = 'red', label = '$O1$')
plt.plot(n,y1, c = 'red')

plt.scatter(n,y2, c = 'blue', label = '$Olog(n)$')
plt.plot(n,y2, c = 'blue')

plt.scatter(n, y3, c = 'green', label = '$O(n)$')
plt.plot(n, y3, c = 'green')

plt.scatter(n, y4, c = 'black', label = '$O(n * log(n))$')
plt.plot(n, y4, c = 'black')

plt.scatter(n, y5, c = 'orange', label = '$O(n^2)$')
plt.plot(n, y5, c = 'orange')

plt.scatter(n, y6, c = 'purple', label = '$O(2^n)$')
plt.plot(n, y6, c = 'purple')

plt.scatter(n, y7, c = 'pink', label = '$O(n!)$')
plt.plot(n, y7, c = 'pink')

plt.scatter(n, y8, c = 'brown', label = '$O(n^n)$')
plt.plot(n, y8, c = 'brown')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='best')

plt.show()
```
